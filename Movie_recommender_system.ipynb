#### Stemming

```python
# ['loving','love','loved'] --> ['love','love','love']
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y= []
    for i in text.split():
        ps.stem()
```

```python
ps.stem('loving'),ps.stem('loved'),ps.stem('dancing'),ps.stem('dance')
```

```python
def stem(text):
    y= []
    for i in text.split():
        y.append(ps.stem(i))
        
    string = " ".join(y)
    return string
```

We will not calculate the euclidean diastance between the vectors, we will calculate the angle between them using cosine distance between the vectors, because euclidean distance is not reliable in higher dimension

```python
from sklearn.metrics.pairwise import cosine_similarity # Ranges between 0-1, more towards 0 more will be similarity
```

```python
cosine_similarity(vectors).shape #angle of each vector with every other vector
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0] # similarity of every movie with every other movie
```

```python
similarity[1]
```

```python
new_df
```

```python
def recommend_5(movie):
    new_df['title'] = new_df['title'].apply(str.lower)
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    # we will sort the distance in descending order for gettling sillimar movies mith more distances, but the problem here is that if we sort our list just based on the distance we may loose the index of that particular distance so we will use the enumertae function here
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    j=1
    for i in movies_list:
        print(j,": ",new_df['title'][i[0]],"\n")
        j+=1
```

```python
def recommend_20_vote_average(vote_average):
    movie_index = new_df[new_df['vote_average']==vote_average].index[0]
    distances = item_similarity[movie_index]
    # we will sort the distance in descending order for gettling sillimar movies mith more distances, but the problem here is that if we sort our list just based on the distance we may loose the index of that particular distance so we will use the enumertae function here
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:21]
    j=1
    arr =[]
    for i in movies_list:
        # print(j,": ",new_df['title'][i[0]],"\n")
        arr.append(new_df['title'][i[0]])
        j+=1
    return arr
```

```python
def find_by_name(title):
    movie_index = new_df[new_df['title']==title].index[0]
    print("Title: ",new_df['title'][movie_index],"||",
          #"vote_count: ",new_df['vote_count'][movie_index],"||",
          "vote_average: ",new_df['vote_average'][movie_index],"||",
         #"popularity: ",new_df['popularity'][movie_index]
         )
```

```python
import math
b = float(input("Enter the average rating: "))
print("\n")
m = float((math.ceil(b)+math.floor(b))/2.0)
if(b>m):
    b = math.ceil(b)
else:
    b = math.floor(b)
recommend_20_vote_average(b)
```

```python
for i in (recommend_20_vote_average(b)):
    find_by_name(i)
    print("\n")
```
